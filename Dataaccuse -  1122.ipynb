```python
import pandas as pd
import numpy as np
import statistics as sta
import warnings
warnings.filterwarnings('ignore')
```

```python
accuse = pd.read_csv('data/accuse.csv')
countriesCode = pd.read_csv('data/countriesCode.csv')

#RENAME COLUMNS
countriesCode.columns = ['Country', 'CountryCode']
accuse.rename(columns = {'SQLDATE':'Date','Actor1CountryCode':'CountryCode','SOURCEURL':'SourceURL','avgTone':'AvgTone'}, inplace=True)

#REMOVE SPACE
countriesCode['CountryCode'] = countriesCode['CountryCode'].str.strip()
countriesCode['Country'] = countriesCode['Country'].str.strip()

accuse
```

```python
#drop data duplicated
accuseDropDup = accuse.drop_duplicates()
accuseDropDup
```

## MERGE DATA WITH COUNTRY CODE

```python
accuseData = accuseDropDup.merge(countriesCode, on='CountryCode', how='left')
accuseData.info()
accuseData
```

## MERGE DATA WITH ACTOR CODE

```python
actorCodeData = pd.read_csv('data/actorCode.csv', delimiter = ',', encoding = "ISO-8859-1")
#RENAME COLUMNS
actorCodeData.columns = ['actorCode','Actor']
#REMOVE SPACE
actorCodeData['actorCode'] = actorCodeData['actorCode'].str.strip()
actorCodeData['Actor'] = actorCodeData['Actor'].str.strip()

accuseDatan = accuseData.merge(actorCodeData, left_on='CountryCode', right_on='actorCode', how='left')
accuseDatan
```

```python
#untuk cek pengaruh actorCode dg Country. cek actornya
accuseDatan[(accuseDatan.Actor.notnull())&(accuseDatan.Country.isnull())]
```

# DEVELOPING/DEVELOPED COUNTRY

```python
devC = pd.read_csv('data/developingCountries.csv', delimiter = ',', encoding = "ISO-8859-1")
devC['CountryCat'] = 'developing'
devC.drop('Country', axis =1, inplace =True)
devC.dropna(inplace=True)
devC
```

```python
accuse = accuseData.merge(devC, on=['CountryCode'], how='left')
accuse.loc[accuse.CountryCode.notnull(), 'CountryCat'] = accuse.loc[accuse.CountryCode.notnull(), 'CountryCat'].fillna('developed')
accuse['HRTopic'] = 'accuse'
accuse = accuse[['Date','Country','CountryCode','CountryCat','HRTopic','AvgTone', 'SourceURL']]
```

```python
accuseDropNA = accuse.dropna()
print('jumlah data demo =', len(accuseDropNA))
accuseDropNA
```
